```python
import matplotlib.pylab as plt
import numpy as np
import csv
```

```python
L = 200

with open(f'./out/tfim_L{L}.csv',newline='')as csvfile:
    reader = csv.DictReader(csvfile)
    bds = []
    es = []
    for row in reader:
        print(row)
        bds.append(float(row['bonddims']))
        es.append(float(row['energies']))
```

```python
plt.figure(figsize=(12,6))
plt.plot(bds,es)
plt.scatter(bds,es)
plt.xticks(bds)
plt.xlabel('bond dimension')
plt.ylabel('E/N')
plt.title('L=200 1D TFIM')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(bds,es)
plt.scatter(bds,es)
plt.xticks(bds)
plt.xlabel('bond dimension')
plt.ylabel('E/N')
plt.title('L=200 AFM Heisenberg chain')
plt.show()
```

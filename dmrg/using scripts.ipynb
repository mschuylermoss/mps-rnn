Look at how different the enlarged MPS energies are from DMRG Energy

```julia
for b in 2:10
    println("Running growing algo")
    chi = 5
    _,e_enlarged = grow_MPS("Heisenberg","open",chi,10,10*b,false);
    println("Running DMRG for small chi")
    _,_,e_dmrg = run_dmrg_1d("Heisenberg", "open", 10*b, chi, false);
    println("Running DMRG")
    _,_,e_exact = run_dmrg_1d("Heisenberg", "open", 10*b, 100, false);
    println("----------------------")
    println("L = $(10*b)")
    println("Energy from L = 10 MPS: ",e_enlarged)
    println("Energy from DMRG with chi = $chi: ",e_dmrg*10*b)
    println("Energy from DMRG: ",e_exact*10*b)
    println("----------------------")
end
```
